```python
import numpy as np
from sklearn.datasets import load_boston
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
```

# load data

```python
#dataSet
X=load_boston().data
X.shape
```

```python
#testSet
y=load_boston().target
y.shape
```

# set model and param_dict 

```python
#param_dict
param={"alpha":range(50,200)}
```

```python
#model
ridge=Ridge(alpha=1)
```

```python
grid=GridSearchCV(estimator=ridge,param_grid=param,scoring="neg_mean_squared_error",cv=10)
```

```python
grid.fit(X,y)
```

# Selection

```python

grid.best_score_
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_index_
```

```python
grid.cv_results_
```

```python
cv_frame=pd.DataFrame(data=grid.cv_results_)
cv_frame
```

```python
plt.plot(range(50,200),-1*cv_frame["mean_test_score"])
plt.show()
```

```python
plt.plot(range(50,200),-1*cv_frame["mean_train_score"])
plt.show()
```

```python
plt.plot(range(50,200),-1*cv_frame["mean_test_score"])
plt.show()
```
